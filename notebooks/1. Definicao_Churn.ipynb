```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../dataset/transaction_new.csv')
df['created_at'] = pd.to_datetime(df['created_at'], format="%Y-%m-%d").dt.tz_localize(None).dt.normalize()
df
```

```python
df_churn = df[['created_at', 'customer_id']]
df_churn = df_churn.sort_values(['created_at'])
df_churn.drop_duplicates(inplace=True)
df_churn
```

Vou remover customer_id duplicados no mesmo dia, para que quando calcular a frequencia eu conte dias distintos

# 2. Calculando a Frequência

```python
df_freq = df_churn.groupby('customer_id').created_at.count().reset_index()
df_freq.columns = ['customer_id','frequency']
df_freq
```

```python
df_churn_filter = df_churn[df_churn['customer_id'].isin(df_freq['customer_id'].values)]
df_churn_filter
```

# Calculando Janela de Tempo para Churn

```python
df_churn_filter['next_date'] = df_churn_filter.groupby('customer_id')['created_at'].shift(-1)
df_churn_filter.fillna({'next_date':pd.to_datetime('2022-08-01', format="%Y-%m-%d")}, inplace=True)
df_churn_filter['lag_date'] = (df_churn_filter['next_date'] - df_churn_filter['created_at']).dt.days

df_churn_filter
```

Como existe customers que não tem registrado a próxima compra vou excluir-los do calculo de janela de tempo, poderia assumir como a data mais recente a próxima compra para esse customer, mas não é garantia de ser real. Mas não custa nada testar depois fazendo esse input

```python
df['created_at'].max()
```

```python
print(f"Máximo: {np.max(df_churn_filter['lag_date'].values)}")
print(f"Minimo: {np.min(df_churn_filter['lag_date'].values)}")
print(f"Média: {np.mean(df_churn_filter['lag_date'].values)}")
print(f"Mediana: {np.median(df_churn_filter['lag_date'].values)}")
print(f"Percentil 20%: {np.percentile(df_churn_filter['lag_date'].values, 25)}")
print(f"Percentil 75%: {np.percentile(df_churn_filter['lag_date'].values, 75)}")
print(f"Percentil 90%: {np.percentile(df_churn_filter['lag_date'].values, 90)}")
print(f"Percentil 95%: {np.percentile(df_churn_filter['lag_date'].values, 95)}")
print(f"Percentil 99%: {np.percentile(df_churn_filter['lag_date'].values, 99)}")
```

Vou assumir como 22 dias sem compras como janela de tempo para considerar Churn

```python
df_labeled = pd.merge(df, df_churn_filter, on=['customer_id', 'created_at'])
df_labeled['churn'] = np.where(df_labeled['lag_date'] >= 22, 1, 0)
df_labeled = df_labeled[['created_at', 'customer_id', 'churn']]
df_labeled
```

O campo churn igual a 1 significa que apôs essa compra o usuário churnou e quando igual a 0 significa que ele não churnou
